```python
import os
import shutil
import subprocess

def move_files_into_folders(source_dir):
    # .jpg
    files = [f for f in os.listdir(source_dir) if f.endswith('.png')]
    print(len(files))

    if not files:
        print("No .jpg files found.")
        return

    # 
    folder_num = 0
    file_count = 0

    for num, f in enumerate(files):
        
        print(num)
        # print(current_folder)
        
        # 
        if num % 1000 == 0:
            folder_num += 1
            current_folder = os.path.join(source_dir, f'folder_{folder_num}')
            print(current_folder)
            if not os.path.exists(current_folder):
                # os.mkdir(current_folder)
                subprocess.run(["mkdir",f"{current_folder}"])
                print("mkdir",f"{current_folder}")
            
        
        # 
        # shutil.move(os.path.join(source_dir, f), current_folder)
        subprocess.run(["mv",f"{source_dir}/{f}",f"{current_folder}/"])
        print("mv",f"{source_dir}/{f}",f"{current_folder}/")
        
        
    #     file_count += 1

    # print(f"Moved {file_count} files into {folder_num - 1} folders.")

if __name__ == "__main__":
    move_files_into_folders("/root/dataset/soumikrakshit-anime-faces/data")
```
