```python
import os

def list_files(directory):
    files = []
    with os.scandir(directory) as entries:
        for entry in entries:
            if entry.is_file():
                files.append(f"{entry.name}")
    return files
                # print(entry.name)

directory = "/root/dataset/High-ResolutionAnimeFaceDataset_512/portraits/folder_1"

all_files = list_files(f"{directory}")
all_files[0:5]
```

```python
import cv2
import numpy as np
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import os

save_image_path = "/root/dataset/High-ResolutionAnimeFaceDataset_512/lline_three_pix/folder_1"

if not os.path.exists(save_image_path):
    os.makedirs(save_image_path)


# for name in all_files:
# 画像の読み込み
# img = cv2.imread(f'{directory}/{name}', cv2.IMREAD_COLOR)
img = cv2.imread(f'/root/source/image_convet/mizore.png', cv2.IMREAD_COLOR)


# グレイスケールに変換
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Cannyエッジ検出
edges = cv2.Canny(gray, 50, 150)


# 線を太くするためのダイレーション
dilated = cv2.dilate(edges, np.ones((2,2), np.uint8))
inverted_edges = cv2.bitwise_not(dilated)
# plt.imshow(edges,gray=True)
# plt.show()
# plt.close()

# 結果を保存
# cv2.imwrite(f'{save_image_path}/{name}', inverted_edges)
cv2.imwrite('/root/source/image_convet/mizore_3.png', inverted_edges)
```
