```python
import cv2
import matplotlib.pyplot as plt

# テキストファイルのパス
# text_file_path = 'wider_face_train_bbx_gt.txt'
text_file_path = '/root/dataset/WiderFaceDataset/wider_face_split/wider_face_train_bbx_gt.txt'

# テキストファイルを開き、内容を読み込む
with open(text_file_path, 'r') as f:
    lines = f.readlines()

# 各行を処理する
for line in lines:
    print("line",line)
    data = line.strip().split()
    print("data",data)
    if len(data) > 1:  # バウンディングボックスの情報がある行のみ処理
        # 画像のパスを取得（注：パスは適宜調整してください）
        img_path = data[0]
        print("img_path",img_path)
        image_full_path = f"/root/dataset/WiderFaceDataset/WIDER_train/images/{data}"
        print("image_full_path",image_full_path)

        # バウンディングボックスの情報を取得
        num_boxes = int(data[1])
        boxes = []
        for i in range(num_boxes):
            box_info = data[2 + i * 10: 2 + (i+1) * 10]
            x, y, w, h = map(int, box_info[:4])  # x, y, width, height
            boxes.append((x, y, w, h))

        # 画像を読み込む
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCVはBGRを使用するため、RGBに変換

        # 各バウンディングボックスに対して矩形を描く
        for (x, y, w, h) in boxes:
            cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)  # 赤色の矩形を描く

        # 画像を表示
        plt.imshow(img)
        plt.show()
```
