```python
import pandas as pd
import time
from PIL import Image
from IPython.display import clear_output
import matplotlib.pyplot as plt

target_dir = "radiantLightV2_1"

dataset_dir = f"/root/dataset/SD_generated_dataset_checked/{target_dir}"
annotated_images = f"{dataset_dir}/images"
annotated_csv = f"{dataset_dir}/annotations/annotations.csv"

annotated_data = pd.read_csv(annotated_csv).values.tolist()

image_num = 18
img = Image.open(f"{annotated_images}/{annotated_data[image_num][0]}")
fig, ax = plt.subplots(1)
ax.imshow(img)
x = annotated_data[image_num][3::2]
y = annotated_data[image_num][4::2]
plt.plot(x, y, 'o',color='red',ms=2)
for i in range(98):
    ax.annotate(i, (x[i], y[i]),color="blue",fontsize = 7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Coordinates')
plt.grid(True)
# plt.gca().invert_yaxis()
plt.show()
time.sleep(1)
clear_output(True)
```

```python
def output_circle_json():
    circle_json = []

    for point_num in range(98):
        if 0 <= point_num <= 32:
            color = "#fa2f2f"
        if 33 <= point_num <= 41:
            color = "#fa802f"
        if 42 <= point_num <= 50:
            color = "#fae62f"
        if 42 <= point_num <= 50:
            color = "#7dfa2f"
        if 51 <= point_num <= 54:
            color = "#2ffa79"
        if 55 <= point_num <= 59:
            color = "#2ffadc"
        if 60 <= point_num <= 67:
            color = "#2fb0fa"
        if 68 <= point_num <= 75:
            color = "#2f4afa"
        if 76 <= point_num <= 87:
            color = "#8e2ffa"
        if 88 <= point_num <= 95:
            color = "#d12ffa"
            
        circle_json_data = {
            "name": f"{point_num + 1}",
            "color": f"{color}",
            "attributes": [],
            "type": "points"
        }
        circle_json.append(circle_json_data)
    return circle_json
circles = output_circle_json()
circles
```

```python
# projectxml
def output_circle_xml(x,y,point_num):
    circle_data_num = f"""<circle r=\"1.5\" stroke=\"black\" fill=\"#b3b3b3\" cx=\"{x}\" cy=\"{y}\" stroke-width=\"0.1\" data-type=\"element node\" data-element-id=\"{point_num}\" data-node-id=\"{point_num}\" data-label-name=\"{point_num}\"></circle>"""
    return circle_data_num.replace("\n","")

import pandas as pd
import time
from PIL import Image
from IPython.display import clear_output
import matplotlib.pyplot as plt

target_dir = "radiantLightV2_1"

dataset_dir = f"/root/dataset/SD_generated_dataset_checked/{target_dir}"
annotated_images = f"{dataset_dir}/images"
annotated_csv = f"{dataset_dir}/annotations/annotations.csv"

annotated_data = pd.read_csv(annotated_csv).values.tolist()

image_num = 18
x = annotated_data[image_num][3::2]
y = annotated_data[image_num][4::2]

xml_sentence = ""
for point_num in range(98):
    xml = output_circle_xml(x[point_num]/512 * 100,y[point_num]/512*100,point_num+1)
    xml_sentence += xml
print(xml_sentence)
```
