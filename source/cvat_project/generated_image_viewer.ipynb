```python
import pandas as pd

target_dir = "anything-v3-fp32"

dataset_dir = f"/root/dataset/SD_generated_dataset_checked/{target_dir}"
annotated_images = f"{dataset_dir}/images"
annotated_csv = f"{dataset_dir}/annotations/annotations.csv"

annotated_data = pd.read_csv(annotated_csv).values.tolist()

# for i in annotated_data:
#     print(i)
```

```python
f"{annotated_images}/{annotated_data[0][0]}"
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(f"{annotated_images}/{annotated_data[0][0]}")
fig, ax = plt.subplots(1)
ax.imshow(img)
```

```python
import time
from IPython.display import clear_output

image_num = 1
for image_num in range(50):
    print(image_num)
    img = Image.open(f"{annotated_images}/{annotated_data[image_num][0]}")
    fig, ax = plt.subplots(1)
    ax.imshow(img)
    x = annotated_data[image_num][3::2]
    y = annotated_data[image_num][4::2]
    plt.plot(x, y, 'o',color='red',ms=2)
    for i in range(98):
        ax.annotate(i, (x[i], y[i]),color="blue",fontsize = 7)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{image_num}')
    plt.grid(True)
    # plt.gca().invert_yaxis()
    plt.show()
    time.sleep(1)
    clear_output(True)
    
```

```python
image_num = 24
img = Image.open(f"{annotated_images}/{annotated_data[image_num][0]}")
fig, ax = plt.subplots(1)
ax.imshow(img)
x = annotated_data[image_num][3::2]
y = annotated_data[image_num][4::2]
plt.plot(x, y, 'o',color='red',ms=2)
for i in range(98):
    ax.annotate(i+1, (x[i], y[i]),color="blue",fontsize = 7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Coordinates')
plt.grid(True)
# plt.gca().invert_yaxis()
plt.show()
time.sleep(1)
clear_output(True)
```
